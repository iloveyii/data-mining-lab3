```python
from sklearn.cluster import KMeans

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt

import time
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
# Global variables
colors = ['green', 'blue', 'orange', 'red', 'purple', 'yellow', 'cyan', 
          'black', 'pink', 'teal', 'aqua', 'maroon']
markers = ["o","v","^","<",">","1","2","3","4","8","s","p","P","*","h","H","+","x","X","D","d"]
dfs = []
df = pd.DataFrame()
file_name = 'employees.csv'
```

```python
def pre_process_data():
    global df
    # Read csv file
    df = pd.read_csv(file_name, delim_whitespace=True)
```

```python
pre_process_data()
```

```python
df
```

```python
target = df.status
inputs = df.drop('status', axis='columns')
```

```python
dummies = pd.get_dummies(inputs.Department)
inputs = pd.concat([inputs, dummies], axis='columns')
inputs.drop('Department', axis='columns', inplace=True)
```

```python
dummies = pd.get_dummies(inputs.age)
inputs = pd.concat([inputs, dummies], axis='columns')
inputs.drop('age', axis='columns', inplace=True)
```

```python
dummies = pd.get_dummies(inputs.salary)
inputs = pd.concat([inputs, dummies], axis='columns')
inputs.drop('salary', axis='columns', inplace=True)
```

```python
inputs
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
